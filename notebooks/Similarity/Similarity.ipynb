```python
class SampleSimilarity(InferenceWorker):
    def __call__(self, t, query, docs_to_compare):
        if t == "text":
            yield "sim_scores", self.sim_from_texts(query, docs_to_compare), "Similarity"
        else:
            yield None        
    def sim_from_texts(self, query, docs_to_compare):
        query_emb = model.encode(query)
        doc_emb = model.encode(docs_to_compare)
        scores = util.dot_score(query_emb, doc_emb)[0].cpu().tolist()
        doc_score_pairs = list(zip(docs_to_compare, scores))
        doc_score_pairs = sorted(doc_score_pairs, key=lambda x: x[1], reverse=True)
        return doc_score_pairs
        
```

```python
worker = SampleSimilarity()
worker.serialize()
```
